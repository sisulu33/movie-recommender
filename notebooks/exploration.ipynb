```python
import pandas as pd

movies = pd.read_csv('../data/movies.csv')
ratings = pd.read_csv('../data/ratings.csv')

print(movies.shape)
print(ratings.shape)
```

```python
movies.head()
```

```python
ratings.head()
```

```python
print("Total movies:", movies['movieId'].nunique())
print("Total users:", ratings['userId'].nunique())
print("Total ratings:", len(ratings))
print("Rating scale:", ratings['rating'].min(), "to", ratings['rating'].max())
```

```python
top_movies = ratings.groupby('movieId')['rating'].count().sort_values(ascending=False).head(10)
top_movies = top_movies.reset_index()
top_movies.columns = ['movieId', 'rating_count']
top_movies.merge(movies, on='movieId')[['title', 'rating_count']]
```
